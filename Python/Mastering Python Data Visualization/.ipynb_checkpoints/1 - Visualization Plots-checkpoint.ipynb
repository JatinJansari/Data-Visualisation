# Visualisation Plots
When creating visualizations, the first step is to be clear on the question to be answered. In other words, how is visualization going to help? There is another challenge that follows this - knowing the right plotting method. Some visualization methods are as follows:
- Bar graph and pie chart
- Box plot
- Bubble chart
- Histogram
- Kernel Density Estimation (KDE) plot
- Line and surface plot
- Network graph plot
- Scatter plot
- Tree map
- Violin plot

In the course of identifying the message that the visualization should convey, it makes sense to look at the following questions:
- How many variables are we dealing with, and what are we trying to plot?
- What do the axis refer to?
- Are the data sizes normalized and does the size of data points mean anything?
- Are we using the right choices of colors?
- For time series data, are we trying to identify a trend or a correlation?

If there are too many variables, it makes sense to draw multiple instances of the same plot on different subsets of the data. This technique is called **lattics** or **trellis** plotting. It allows a viewer to quickly extract a huge amount of information about complex data.

# Starting point

Most visualization stories begin with some question that is oriented towards a topic where the data is being either explored or collected. The question contains the premise to the story and leads us to the point at which the data takes an expedition over the storyline.

The key parts of the story relate to the process of placing the visualization in a meaningful context. The context provides knowledge that answers questions such as the following:
- Is there sufficient data?
- Is there a time frame within which this data exists?
- Which associable events around the globe will influence this data?

## The process
1. Acquiring, parsing and filtering data to detect outliers and abnormalities, data mining and refining, visual representation, and interaction.
2. Creating interesting stories with data
3. Perception, presentation methods, and best practice for visualization
4. Interactive visualization - exploring event listeners and layouts.

# Example: Ebola 2014 data
### **Question:** How many Ebola deaths were reported in the year 2014?

We will use specific data from the World Health Organization (http://www.who.int/en/). The data contains information about the spread of the Ebola disease in Guinea, Liberia, Mali, Nigeria, Senegal, Sierra Leone, Spain, United Kingdom, and the United States of America.

The URL for this information is: https://data.hdx.rwlabs.org/dataset/ebola-cases-2014/

## Data
The contents of the data file in the CSV format include the indicator, country name, date, and the number of deaths or the number of infections depending upon what the indicator says. There are 36 different indicators.

```python
import pandas as pd
eboladata = pd.read_csv("ebola.csv")
eboladata
```

```python
filtered = eboladata[eboladata["value"]>0]
filtered = filtered[filtered["Indicator"].str.contains("deaths")]
filtered
```

```python
len(filtered)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import csv
import datetime as dt

guineadata = filtered[(filtered['Country'] == "Guinea") & (filtered['Indicator'] == "Cumulative number of confirmed Ebola deaths")]
sierradata = filtered[(filtered['Country'] == "Sierra Leone") & (filtered['Indicator'] == "Cumulative number of confirmed Ebola deaths")]
liberiadata = filtered[(filtered['Country'] == "Liberia") & (filtered['Indicator'] == "Cumulative number of confirmed Ebola deaths")]
```

```python
g_x = [dt.datetime.strptime(row, '%Y-%m-%d').date() for row in guineadata['Date']]
g_y = guineadata['value']

s_x = [dt.datetime.strptime(row, '%Y-%m-%d').date() for row in sierradata['Date']]
s_y = sierradata['value']

l_x = [dt.datetime.strptime(row, '%Y-%m-%d').date() for row in liberiadata['Date']]
l_y = liberiadata['value']
```

```python
plt.figure(figsize = (10,10))
plt.plot(g_x, g_y, color = 'red', linewidth = 2, label = "Guinea")
plt.plot(s_x, s_y, color = 'blue', linewidth = 2, label = "Sierra Leone")
plt.plot(l_x, l_y, color = 'orange', linewidth = 2, label = "Liberia")
plt.xlabel('Data', fontsize = 18)
plt.ylabel('Number of Ebola Deaths', fontsize = 18)
plt.title("Confirmed Ebola Deaths", fontsize = 20)
plt.legend(loc = 2)
plt.show()
```

# Example: Sports
### Question: What are the top five record touchdowns by quarterbacks in American Footbal as of Feb 2015?

Data source: http://www.pro-football-reference.com/players/D/DawsLe00.htm

## Data
The data contains information about the top 22 quarterbacks. These quarterbacks played in different time periods.

We will use **``csv``** module as it has some convenient functions such as **``DictReader``**. With **``DictReader``** and **``DictWriter``** the rows of csv data are translated to dictionaries with field anmes instead of lists.

```python
import csv
import matplotlib.pyplot as plt

with open('qb_data.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        name = row['Name']
        tds = row['TD']
```

In order to convert some of the values to a number, we may need a function that converts and returns a numeric value.

```python
# num(s) and getcolors() functions
def num(s):
    try:
        return int(s)
    except ValueError:
        return 0

def getcolors():
    colors = [(31, 119, 180), (255,0,0), (0,255,0), (148, 103, 189), (140, 86, 75), (218, 73, 174), (127, 127, 127), (140,140,26), (23, 190, 207), (65,200,100), (200, 65,100), (125,255,32), (32,32,198), (255,191,201), (172,191,201), (0,128,0), (244,130,150), (255, 127, 14), (128,128,0), (10,10,10), (44, 160, 44), (214, 39, 40), (206,206,216)]
    
    for i in range(len(colors)):
        r, g, b = colors[i]
        colors[i] = (r/255.0, g/255.0, b/255.0)
    return colors
```

```python
def getQbNames():
    qbnames = ['Peyton Manning']
    blankname = ''
    i = 0
    with open('qb_data.csv') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            if ( row['Name'] != blankname and qbnames[i] != row['Name']):
                qbnames.append(row['Name'])
                i += 1
    return qbnames

def readQbdata():
    resultdata = []
    with open('qb_data.csv') as csvfile:
        reader = csv.DictReader(csvfile)
        resultdata = [row for row in reader]
    return resultdata
```

```python
fdata = []
prevysum = 0

qbnames = getQbNames()
fdata = readQbdata()
```

```python
i = 0
rank = 0
prevysum = 0
lastyr = 0
highrank = 247
colorsdata = getcolors()

fig = plt.figure(figsize = (15, 13))
ax = fig.add_subplot(111, axisbg = 'white')

# limits for TD
plt.ylim(10, 744)
plt.xlim(1940, 2021)

colindex = 0
lastage = 20

for qbn in qbnames:
    x = []
    y = []
    prevysum = 0
    for row in fdata:
        if ( row['Name'] == qbn and row['Year'] != 'Career'):
            yrval = num(row['Year'])
            lastage = num(row['Age'])
            prevysum += num(row['TD'])
            lastyr = yrval
            x += [yrval]
            y += [prevysum]
            
    if (rank > highrank):
        plt.plot(x, y, color = colorsdata[colindex], label = qbn, linewidth = 2.5)
        plt.legend(loc = 0, prop = {'size': 10})
        colindex = (colindex + 1)%22
        plt.text(lastyr - 1, prevysum + 2, qbn + "(" + str(prevysum) + "): " + str(lastage), fontsize = 9)
    
    else:
        plt.plot(x, y, color = colorsdata[22], linewidth = 1.5)
        rank += 1

plt.xlabel('Year', fontsize = 18)
plt.ylabel('Cumulative Touch Downs', fontsize = 18)
plt.title("Cumulative Touch Downs by Quarter Backs", fontsize = 20)
plt.show()
```
